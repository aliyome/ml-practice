```python
import datetime

import numpy as np
import pandas as pd

import plotly
import plotly.express as px
import plotly.figure_factory as ff
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots
from plotly.graph_objs.scatter import Line
from plotly.graph_objs import Scatter
```

```python
data = np.loadtxt('ex1data2.txt', delimiter=',')
data
```

```python
X = data[:, [0, 1]]
y = np.array([data[:, 2]]).T
X.shape, y.shape
```

```python
def feature_normalize(X):
    mu = np.mean(X, 0)
    sigma = np.std(X, 0)
    X_norm = (X - mu) / sigma
    return X_norm, mu, sigma
```

```python
XX, mu, sigma = feature_normalize(X)
XX, mu, sigma
```

```python
alpha = 0.01
num_iters = 400
theta = np.zeros((3, 1))
theta
```

```python
def h(X, theta):
    return X @ theta
```

```python
def computeCost(X, y, theta):
    m =  X.shape[0]
    J = 1/(2 * m) * np.sum((h(X, theta) - y) ** 2)
    return J
```

```python
def gradientDescentMulti(X, y, theta, alpha, num_iters):
    m = X.shape[0]
    theta_out = theta
    J_history = np.zeros(num_iters)
    for i in range(num_iters):
        theta_out -= alpha / m * X.T @ (h(X, theta_out) - y)
        J_history[i] = computeCost(X, y, theta_out)
    return theta_out, J_history
```

```python
theta = np.zeros((3, 1))
XwithX0 = np.c_[np.ones((XX.shape[0], 1)), XX]
th, J_history = gradientDescentMulti(XwithX0, y, theta, alpha, num_iters)
J_history
```

```python
px.scatter(y=J_history)
```

```python
params = np.array([1650, 3])
params = (params - mu) / sigma
params = np.r_[1, params]
price = params @ th
th, params, price
```
