```python
sample1 = ''
with open("emailSample1.txt", 'r') as fp:
    sample1 = fp.read()

print(sample1)

vocab = pd.read_csv('vocab.txt', delimiter='\t', header=None)
vocab = vocab.set_index(1)
vocab
```

```python
import re

a = sample1
a = a.lower()
a = re.sub(r"<[^<>]+>", " ", a)
a = re.sub(r"\d+", 'number', a)
a = re.sub(r"(http|https)://[^\s]+", 'httpaddr', a)
a = re.sub(r"[^\s]+@[^\s]+", 'emailaddr', a)
a = re.sub(r"[\$]+", 'dollar', a)
a = re.sub(r"[\'@$\/#.-:&*+=\[\]?!(){},\">_<;%\n]", ' ', a)
a = re.sub(r"\s+", ' ', a)
a = re.sub(r"(^\s+|\s+$)", "", a)

print(a)
words = a.split(' ')
words
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()

vocab['count'] = 0

w = stemmer.stem(words[0])
for w in words:
    w = stemmer.stem(w)
    if w in vocab.index:
        vocab.loc[w]['count'] += 1

vocab['count'][vocab['count'] == 0] = np.nan
vocab.dropna()
```

```python
data = loadmat('spamTrain.mat')
X = data['X']
y = data['y']

data = loadmat('spamTest.mat')
Xtest = data['Xtest']
ytest = data['ytest']
```
